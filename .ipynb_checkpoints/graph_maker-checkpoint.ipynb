```python
import math
import matplotlib.pyplot as plt
from matplotlib.patches import Wedge, Circle

def gauge_percent(value,
                  vmin=-1.0, vmax=0.0,           # range as fractions (-1 = -100%, 0 = 0%)
                  marker=None,                   # optional slim marker line
                  title="MaxDrawdown_%", 
                  as_percent=True,
                  color=["556B2F","8FA31E","C6D870","EFF5D2"]):
    """
    Draw a semi-circular gauge for a percentage-like metric.

    value:      current value (fraction, e.g., -0.39 for -39%). If you pass -39, set as_percent=False.
    vmin/vmax:  min/max of the scale (fractions).
    marker:     draw a slim radial marker at this value (fraction). Set to None to disable.
    title:      text above the gauge.
    as_percent: if False, interpret inputs as whole % (e.g., -39 instead of -0.39).
    color:      arc + needle color.
    """
    Cprimary, Csecond, Cthird,Cfourth =  color



    
    if not as_percent:
        value   = value   / 100.0
        vmin    = vmin    / 100.0
        vmax    = vmax    / 100.0
        marker  = None if marker is None else marker / 100.0

    # clamp value
    value = min(max(value, vmin), vmax)

    def val2ang(v):
        """map value in [vmin,vmax] → angle in degrees [180,0]."""
        t = (v - vmin) / (vmax - vmin)
        return 180 * (1 - t)

    fig, ax = plt.subplots(figsize=(3.8, 2.6), subplot_kw={"aspect": "equal"})
    ax.set_xlim(-1.2, 1.2)
    ax.set_ylim(-0.1, 1.2)
    ax.axis("off")

    # background
    base = Wedge((0,0), r=1.0, theta1=0, theta2=180, width=0.25,
                 facecolor="#EFF5D2", edgecolor="none")
    ax.add_patch(base)

    # main arc
    theta2 = val2ang(value)
    arc = Wedge((0,0), r=1.0, theta1=theta2, theta2=180, width=0.25,
                facecolor=color, edgecolor="none")
    ax.add_patch(arc)

    # needle
    ang = math.radians(val2ang(value))
    ax.plot([0.15*math.cos(ang), 1.05*math.cos(ang)],
            [0.15*math.sin(ang), 1.05*math.sin(ang)],
            lw=3, color=color)
    ax.add_patch(Circle((0,0), radius=0.04, color=color))

    # optional marker
    if marker is not None:
        ma = math.radians(val2ang(marker))
        # marker line
        ax.plot([0.78*math.cos(ma), 1.12*math.cos(ma)],
                [0.78*math.sin(ma), 1.12*math.sin(ma)],
                lw=2, color="red")
        # label slightly outside the arc
        ax.text(1.25*math.cos(ma), 1.25*math.sin(ma),
                f"{marker:.0%}", ha="center", va="center",
                fontsize=9, color="red")

    # ticks (min and max)
    for tv in [vmin, vmax]:
        lab = f"{tv:.0%}"
        ta = math.radians(val2ang(tv))
        ax.text(1.12*math.cos(ta), 1.12*math.sin(ta), lab,
                ha="center", va="center", fontsize=9, color="#6b6b6b")

    # big center text
    ax.text(0, 0.35, f"{value:.0%}", ha="center", va="center", fontsize=26, color="#6b6b6b")
    ax.text(0, 1.05, title, ha="center", va="bottom", fontsize=11)

    plt.tight_layout()
    return fig, ax

# Example
if __name__ == "__main__":
    # value = -39%, marker = -15%
    fig, ax = gauge_percent(-0.39, vmin=-1.0, vmax=0.0, marker=-0.15, title="MaxDrawdown_%")
    plt.show()
```
